```python
# Importing Libraries

import pandas as pd
import numpy as np
```

```python
# Creating a random array of size 10 rows and 4 columns

arr1 = np.random.randint(100, size = (10,4))
```

```python
# Creating a Pandas dataframe

df = pd.DataFrame()
df[['Col1', 'Col2', 'Col3', 'Col4']] = arr1
```

```python
df.head()
```

```python
# Renaming the columns

df = df.rename(columns = {'Col1':'Random value 1', 'Col2':'Random value 2', 'Col3':'Random value 3', 'Col4':'Random value 4'})
```

```python
# Datafram statistics

df.describe()
```

```python
#Checking for Null values

df.isnull()
```
